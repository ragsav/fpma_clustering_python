```python
import string 
import random 
import numpy as np
from sklearn.cluster import KMeans
from nsetools import Nse
# stocks = ["CIPLA","GAIL","BHEL","SAIL","TATA","ACE","BLUEDART"]

risk_option_weights=[[0,2,5,8],[1,3,7,10],[0,1,4,8],[3,6,8,0],[0,4,8,0],[1,3,7,8]]
```

```python
users=[]
riskScores=[]
for x in range(1000):
    user={}
    user_id = ''.join(random.choices(string.ascii_uppercase + string.digits, k = 20))
    riskScore = []
    risk_data = np.random.randint(4, size=6)
    for i,r in enumerate(risk_data):
        riskScore.append(risk_option_weights[i][r])
        
    user["user_id"] = user_id
    user["riskScore"]=riskScore
#     print(riskScore)
    riskScores.append(riskScore)
    users.append(user)

    
print(riskScores)
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(riskScores)
cluster_labels = kmeans.labels_
```

```python
users_with_label=users

for x in range(1000):
    users_with_label[x]["category_label"] = cluster_labels[x]

users_with_label
    
```

```python
nse = Nse()
all_stock_codes = nse.get_stock_codes()
# random.choice(all_stock_codes)
stock_symbols=[];

for s in all_stock_codes:
    stock_symbols.append(s)
    
stock_symbols.pop(0)

stock_symbols = random.choices (stock_symbols,k=50)
stock_symbols
```

```python
for x in range(len(users_with_label)):
    users_with_label[x]["watch_list_symbols"] = random.choices(stock_symbols,k=10)
    print(users_with_label[x])
    
```
